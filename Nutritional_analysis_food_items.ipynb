```python
# Step1: INSTALL LIBRARIES (IF NOT ALREADY INSTALLED)
! pip install pandas matplotlib seaborn

# Step2: IMPORT LIBRARIES 
import pandas as pd  # type: ignore
import matplotlib.pyplot as plt  # type: ignore
import seaborn as sns # type: ignore

# Step3: Create data 
data= {
    "Food": ["Apple", "Banana", "Orange", "Broccoli", "Carrot", "Spinach", "Chicken", "Salmon", "Egg", "Almonds", "Walnuts", "Rice", "Oats", "Cheese", "Yogurt", "Bread", "Butter", "Potato", "Tomato", "Peanut Butter"],
    "Category": ["Fruit" , "Fruit", "Fruit", "Vegetable", "Vegetable", "Vegetable", "Meat", "Fish", "Protein", "Nuts", "Nuts", "Grain", "Grain", "Dairy", "Dairy", "Grain", "Dairy", "Vegetable", "Vegetable", "Nuts"],
    "Protein":[0.3,1.3,0.9,2.8,0.9,2.9,31,20,13,21,15,2.6,17,25,10,9,0.9,2,0.9,25],
    "Fat":[0.2,0.3,0.1,0.4,0.2,0.4,3.6,13,11,49,65,0.9,7,33,0.4,3.2,81,0.1,0.2,50],
    "Carbs":[14,27,12,7,10,3.6,0,0,1.1,22,14,23,66,1.3,4.7,49,0.1,17,3.9,20],
    "Sugar":[10,14,9,1.5,5,0.4,0,0,1.1,4.4,2.6,0.4,1,0.5,4.7,5,0.1,0.8,2.6,9],
    "Calories":[52,105,47,34,41,23,165,208,155,576,654,112,389,403,59,265,717,77,18,588]
}

df= pd.DataFrame(data)

# Step4: Look on the data prepared
print(df.head())
print(df.tail())
print(df.info())

# Step5: Data analysis 

## A. Top 5 common foods by protein 
top_protein= df[['Food','Protein']].sort_values(by='Protein', ascending=False).head(5)
top_protein.plot.bar(x='Food', y='Protein', title='TOP 5 COMMON FOODS BY PROTEIN', figsize=(10,6))
plt.show


## B. Top 5 common foods by carbohydrates 
top_carbohydrate= df[['Food', 'Carbs']].sort_values(by='Carbs', ascending=False).head(5)
top_carbohydrate.plot.bar(x='Food',y='Carbs', title='TOP 5 COMMON FOODS BY CARBOHYDRATE', figsize=(10,6),color='red')
plt.show()


## C. Top 5 common foods by fats  
top_fat= df[['Food', 'Fat']].sort_values(by='Fat', ascending=False).head(5)
top_fat.plot.bar(x='Food',y='Fat', title='TOP 5 COMMON FOODS BY FAT', figsize=(10,6),color='green')
plt.show()


## D. Distribution of Macronutrients
sns.histplot(df['Protein'], kde=True, color='blue', label='Protein')
sns.histplot(df['Fat'], kde=True, color='green', label='Fat')
sns.histplot(df['Carbs'], kde=True, color='red', label='Carbs')
plt.legend()
plt.title("DISTRIBUTION OF MACRONUTRIENTS")
plt.show()


## E. Correlation Between Macronutrients 
plt.figure(figsize=(10,6))
sns.heatmap(df[['Protein', 'Fat', 'Carbs']].corr(), annot=True, cmap="BrBG")
plt.title("CORRELATION BETWEEN MACRONUTRIENTS")
plt.show()


# Step6: Save data for later usage
df.to_csv("food_nutrition_data.csv", index=False)
print("GOT THE RESULT SUCCESSFULLY")













```
